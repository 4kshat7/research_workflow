## 2. Show it with a pirate heading + logos

Here we load the saved PNG and draw a banner across the top: the **GitHub**
logo on the left, the **BioConnect** logo on the right, and a pirate heading in
the middle. Edit the `heading` text or the logo paths to taste.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.patheffects as pe
from matplotlib.patches import Rectangle
from pathlib import Path

output_path = Path('..') / 'outputs' / 'wordcloud.png'

# ---- Customise here ----
heading = 'I am a Git Pirate now! √'
left_logo_path = Path('..') / 'images' / 'github-1.jpg'
right_logo_path = Path('..') / 'images' / 'bioconnect_uobd_logo.jpeg'

def load(p):
    """Read an image if it exists, else return None (so a missing logo is skipped)."""
    return mpimg.imread(p) if p.exists() else None

if not output_path.exists():
    print('No word cloud yet - run the cell above first.')
else:
    img = mpimg.imread(output_path)
    h, w = img.shape[0], img.shape[1]

    fig, ax = plt.subplots(figsize=(8, 8))
    ax.imshow(img)
    ax.axis('off')

    # White banner across the top so the logos + text stay readable.
    banner_h = h * 0.16
    ax.add_patch(Rectangle((0, 0), w, banner_h, color='white', alpha=0.9, zorder=3))

    # Each logo is placed in a square box inside the banner.
    logo_size = banner_h * 0.78
    top_y = (banner_h - logo_size) / 2          # smaller y = higher on screen
    bottom_y = top_y + logo_size

    left_logo = load(left_logo_path)
    if left_logo is not None:
        x0 = w * 0.02
        ax.imshow(left_logo, extent=[x0, x0 + logo_size, bottom_y, top_y], zorder=4)

    right_logo = load(right_logo_path)
    if right_logo is not None:
        x1 = w * 0.98
        ax.imshow(right_logo, extent=[x1 - logo_size, x1, bottom_y, top_y], zorder=4)

    # Pirate heading, centred between the two logos - styled like a big title:
    # large serif, bold, dark-green fill with a white outline so it pops.
    ax.text(w * 0.5, banner_h * 0.5, heading, ha='center', va='center',
            fontsize=27, fontweight='bold', fontfamily='serif',
            fontstyle='italic', color="#a50505", zorder=5,
            path_effects=[pe.withStroke(linewidth=3.5, foreground='white')])

    # Lock the view to the word-cloud image (overlay imshows can rescale it).
    ax.set_xlim(0, w)
    ax.set_ylim(h, 0)
    ax.set_aspect('equal')

    plt.tight_layout()
    plt.show()
```

### Tweaks

- **Change the wording:** edit the `heading` variable.
- **Bigger / smaller title:** change `fontsize` on the `ax.text(...)` call.
- **Swap a logo:** point `left_logo_path` / `right_logo_path` at any image in
  `images/`. A missing file is skipped automatically (no crash).
- **Resize / reposition:** adjust `banner_h` (banner height), `logo_size`, and
  the `x0` / `x1` margins.
